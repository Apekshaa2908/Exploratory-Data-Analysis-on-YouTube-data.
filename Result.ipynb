13. Does the number of views have any effect on how many people disliked the video? Support your
answer with a metric and a plot.

```python
df1=df[['view_count','dislikes']]
```

```python
df1.cov()
```

```python
df1.corr()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df1['view_count'],df1['dislikes'])
plt.xlabel('view_count')
plt.ylabel('dislikes')
plt.title('view_count vs dislikes')
plt.show()
```

## view_count has effects on the no.of dislikes. The coorelation is 0.6 and they are directly proportional to each other
